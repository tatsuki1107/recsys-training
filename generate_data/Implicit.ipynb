# How algorithmic confounding in recommendation systems increases homogeneity and decreases utility を参考にしたシミュレーション

https://dl.acm.org/doi/abs/10.1145/3240323.3240370

```python
import sys;sys.path.append('..')
import warnings;warnings.filterwarnings('ignore')
import numpy as np
from scipy.stats import dirichlet
import scipy.stats as stats
from util.plot import plot_heatmap
np.random.seed(12345)

n_users = 100
n_items = 10000
d = 10

mu_p_hat = dirichlet([1 for _ in range(d)]).rvs(n_users)
mu_alpha_hat = dirichlet([100 for _ in range(d)]).rvs(n_items)

mu_p = 10*mu_p_hat
mu_alpha = 0.003*mu_alpha_hat

p_u = np.array([dirichlet.rvs(mu_p[u], size=1).flatten().tolist() for u in range(n_users)])
alpha_i = np.array([dirichlet.rvs(mu_alpha[i], size=1).flatten().tolist() for i in range(n_items)])
mu = np.dot(p_u, alpha_i.T)
var = 1e-5

def res_alpha_and_beta(mu, var=0.00001):
    alpha = (((1-mu)/(var**2))-(1/mu))*(mu**2)
    # アルファとベータが0以下になるのを防ぐ
    alpha = np.where(alpha > 0, alpha, 0.00001)
    beta = alpha*((1/mu)-1)
    beta = np.where(beta > 0, beta, 0.00001)
    
    return alpha, beta

alpha, beta = res_alpha_and_beta(mu)
Vui = stats.beta.rvs(alpha, beta)
```

```python
plot_heatmap(Vui.T)
```
