```python
import pandas as pd
df = pd.read_csv("./data/purchase_history.csv")
```

```python
join_str = lambda i: "2022-"+i
date_list = df["date"].to_list()
date_list = map(join_str, date_list)
df["date"] = list(date_list)
```

```python
df.to_csv("./data/purchase_history.csv", index=False)
```
